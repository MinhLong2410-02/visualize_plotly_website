## Mô tả bộ dữ liệu 
Các thuộc tính:
- `date`: Ngày thu ghi nhận dữ liệu 
- `tmin`: Nhiệt độ thấp nhất
- `tmax`: Nhiệt độ cao nhất
- `prcp`: Lượng mưa
- `snow`: Tuyết rơi (có hay không)
- `snwd`: Độ dày của tuyết
- `awnd`: Tốc độ gió trung bình

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# đọc dữ liệu 
df = pd.read_csv('weather_data.csv',parse_dates=['date'])
```

```python
# xem 10 dòng đầu tiên của bộ dữ liệu 
df.head(10)
```

```python
# xem kích thuốc dữ liệu
df.shape
```

```python
# xem thông tin của dữ liệu 
df.info()
```

```python
# số lượng giá trị null ở các cột
df.isnull().sum()
```

- Ta thấy số lượng giá trị *null* có trong dữ liệu rất ít. Vì vậy, chúng ta sẽ chọn cách xóa bỏ những hàng dữ liệu chứa giá trị *null*

```python
# xóa các hàng có giá trị null 
df = df.dropna() 
```

```python
# kiểm tra lại dữ liệu sau khi xóa giá trị null
df.isnull().sum()
```

```python
# xem tên cột 
df.columns
```

- Chúng ta sẽ thực hiện đổi lại tên cột để dễ hiểu và xử lý hơn

```python
# đổi tên cột
new_name = {'date': 'date', 'tmin':'t_min', 'tmax':'t_max', 'prcp':'precipitation', 
            'snow':'snow_fall', 'snwd':'now_depth', 'awnd':'wind_speed'}
df.rename(columns = new_name, inplace=True)
```

```python
# kiểm tra kết quả
df.head()
```

```python
# Outlier Analysis
fig, axs = plt.subplots(6, figsize = (16,9))
plt1 = sns.boxplot(data = df, x = df['t_min'], ax = axs[0])
plt2 = sns.boxplot(data = df, x = df['t_max'], ax = axs[1])
plt3 = sns.boxplot(data = df, x = df['precipitation'], ax = axs[2])
plt4 = sns.boxplot(data = df, x = df['snow_fall'], ax = axs[3])
plt5 = sns.boxplot(data = df, x = df['snow_fall'], ax = axs[4])
plt6 = sns.boxplot(data = df, x = df['wind_speed'], ax = axs[5])
plt.tight_layout()
```

### Phân tích đa biến 

```python
# Phân phối nhiệt độ thấp nhất và nhiệt độ cao nhất khác nhau như thế nào 
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18, 5))
fig.suptitle('SO SÁNH HISTOGRAM CỦA T_MIN VÀ T_MAX', fontsize=16)
sns.histplot(data=df['t_min'], ax = ax[0]);
plt.xlabel("Temperature Minimum");
sns.histplot(data=df['t_max'], ax = ax[1]);
plt.xlabel("Temperature Maximum");
```

```python
# tất cả các thuộc tính trong dữ liệu có phân phân phối như thế nào ?
fig, ax = plt.subplots(nrows=2, ncols=3, figsize=(20, 10))
fig.suptitle('phân phối của tất cả các thuộc tính', fontsize=16)

sns.histplot(data=df['t_min'], ax = ax[0, 0]);
sns.histplot(data=df['t_max'], ax = ax[0, 1]);
sns.histplot(data=df['precipitation'], ax = ax[0, 2]);
sns.histplot(data=df['snow_fall'], ax = ax[1, 0]);
sns.histplot(data=df['now_depth'], ax = ax[1, 1]);
sns.histplot(data=df['wind_speed'], ax = ax[1, 2]);
```

```python
corr = df.corr()
# ones_like can build a matrix of booleans (True, False) with the same shape as our data
ones_corr = np.ones_like(corr, dtype=bool)
# np's triu: return only upper triangle matrix
mask = np.triu(ones_corr)
adjusted_mask = mask[1:, :-1]
adjusted_corr = corr.iloc[1:, :-1]

fig, ax = plt.subplots(figsize=(10,8))


#That method uses HUSL colors, so you need hue, saturation, and lightness. 
#I used hsluv.org to select the colors of this chart.
cmap = sns.diverging_palette(0, 230, 90, 60, as_cmap=True)

sns.heatmap(data=adjusted_corr, mask=adjusted_mask,
            annot=True, annot_kws={"fontsize":13}, fmt=".2f", cmap=cmap,
            vmin=-1, vmax=1,
            linecolor='white', linewidths=0.5);

yticks = [i.upper() for i in adjusted_corr.index]
xticks = [i.upper() for i in adjusted_corr.columns]

ax.set_yticklabels(yticks, rotation=0, fontsize=13);
ax.set_xticklabels(xticks, rotation=90, fontsize=13);
title = 'TƯƠNG QUAN GIỮA CÁC THUỘC TÍNH\n'
ax.set_title(title, loc='left', fontsize=18);
```
